```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read loss and get the last number from each line

with open('loss.txt', 'r') as f:
    loss = f.read().splitlines()
    loss = [float(i.split(' ')[-1]) for i in loss]
```

```python
num = np.linspace(0,40, 80)

plt.rcParams['figure.figsize'] = [16, 16]
plt.plot(num, loss)
plt.xlabel('Number of epoch', size=20)
plt.ylabel('Loss', size=20)
plt.xticks(size=20)
plt.yticks(size=20)
plt.grid('on')
```
